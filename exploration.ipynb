Distribution of ArrDelay and DepDelay are pretty the same. Delay time has long tail distributions with a lot of flights has delay time less than 20 min and few has long delay time. When plotted on log-scale, data still has right skewed distributions, with 2 peaks on 10 and 20 mins.

## Multivariate Exploration

> The main thing I want to explore in this part of the analysis is how the five reason of delay distribute across other qualitative variables


```python
# The cariers
carriers_order=df_reason.groupby('UniqueCarrier')['ArrDelay'].mean().sort_values(ascending=False).index
reason_order=['LateAircraftDelay','NASDelay','CarrierDelay','WeatherDelay','SecurityDelay']
artists=[]
baselines=np.zeros(len(carriers_order))
carrier_time=df_reason.groupby('UniqueCarrier')['Time'].mean()
for i in range(5):
    reason=reason_order[i]
    inner_time=df_reason[df_reason['Reason']==reason].groupby('UniqueCarrier')['Time'].mean()
    inner_prop=inner_time/carrier_time
    bars=plt.bar(x=np.arange(len(carriers_order)),height=inner_prop[carriers_order], 
                 bottom=baselines)
    artists.append(bars)
    baselines+=inner_prop[carriers_order];
plt.xticks(np.arange(len(carriers_order)), carriers_order)
plt.yticks(np.arange(0,5,1), np.arange(0,1,0.2))
plt.legend(reversed(artists), reversed(reason_order), framealpha = 1,
           bbox_to_anchor = (1, 0.5), loc = 6); 
```

```python
#cluster bar chart
plt.figure(figsize = [12, 5])

sb.barplot(data = df_reason, x = 'UniqueCarrier',y='Time', hue = 'Reason',
             order = carriers_order, hue_order = reason_order)
plt.legend();
```

For most of the carriers, CarrierDelay, NASDelay and LateAircraftDelay are the most often happened reasons. But OH is one exception when having WeatherDelay as the most often reason

Next is to see the change of time delay in 5 reason by date

```python
fig=plt.figure(figsize=[16,10])
delay_by_date=df_reason.groupby(['Month','Reason'])['Time'].mean().reset_index()
sb.pointplot('Month', 'Time', data=delay_by_date, hue='Reason')
plt.show();
```

SecurityDelay and WeatherDelay is quite stable by time, while LateAircraftDelay and NASDelay increase in the middle of the year then fail in Sep and increase again in the end of year, CarrierDelay not follow the same pattern

```python
# Scheduled Arrival Time
plt.figure(figsize=[16,5])

bin_center=np.arange(df_reason['CRSArrTime'].min(), df_reason['CRSArrTime'].max() +100, 100)
arrtime_binned=pd.cut(df_reason['CRSArrTime'], bin_center, include_lowest=True)
sb.barplot( data=df_reason, x=arrtime_binned, y=df_reason['Time'],hue='Reason')
plt.xticks(np.arange(0, 24, 1),np.arange(0, 24, 1)) ;
```

There is an interesting find out here. While LateAircraftDelay happend more in the early and late flight, NASDelay and CarrierDelay have the opposite pattern, which happen more in middle of the day.
