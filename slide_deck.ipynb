## Investigation Overview

In this investigation, I wanted to look at the relationship between flights delay time and characteristics of flights to see if they have any interesting partern. The main focus was on the five reason of Delay: CarrierDelay, WeatherDelay, NASDelay, SecurityDelay, LateAircraftDelay.

## Dataset Overview

The dataset consitst of 7009724 unique records about flights in the United States in 2008, including date of flight, actual and scheduled arrival time, carriers, airport code, arrival and departure delays time, and reasons for delays. 4 duplicated records have been removed and one new atribute - Date in the full format yyyy-MM-DD has been created. The five reasons in seperated columns have been transformed to 2 new atributes: Reason and Time

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df=pd.read_csv('./data/2008.csv')
df.drop_duplicates(keep='first',inplace=True)
df['Date']=pd.to_datetime(dict(year=df.Year, month=df.Month, day=df.DayofMonth))
```

```python
df_cut=df[['FlightNum','TailNum','Month','Date','DepTime','UniqueCarrier','Origin','CRSArrTime','ArrDelay','CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']]
df_reason=df_cut.melt(id_vars=['FlightNum','TailNum','Month','Date','DepTime','UniqueCarrier','Origin','CRSArrTime','ArrDelay'], 
              value_vars=['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay'],
              var_name='Reason', value_name='Time')
df_reason=df_reason.loc[df_reason.Time.notnull()]
```

## Distribution of Arrival Delay Time

> Arrival Delay Time in the dataset take on a very large range of values, from about -500 mins at the lowest, to about 2500 mins at the highest. Plotted on a logarithmic scale, data still has right skewed distributions, with 2 peaks on 10 and 20 min.

```python
log_binsize=0.025
plt.figure(figsize=[16,4])
# Arrival Delay
plt.subplot(1,2,1)
bins=10**np.arange(1, np.log10(df['ArrDelay'].max()) +log_binsize, log_binsize)
plt.hist(data=df, x='ArrDelay', bins=bins)
plt.xscale('log')
plt.xticks([10, 20, 50, 100, 200,500],[10, 20, 50, 100, 200,500])
plt.xlabel('Arrival Delay Time (min)')
plt.title('Distribution of Arrival Delay Time')
;
```

## Distribution of Delay Time for each reason

> There are 5 reasons for the arrival delay time: CarrierDelay, WeatherDelay, NASDelay, SecurityDelay and LateAircraftDelay. Total time of those 5 reasons is equal to the Arrival Delay Time and the distribution patterns are also the same as of Arrival delay time.

```python
x_vars=['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
fig=plt.figure(figsize=[16,10])
binsize=0.025
base_color=sb.color_palette()[0]
for i, var_name in enumerate(x_vars):
    bins=10**np.arange(1, np.log10(df[var_name].max()) +binsize, binsize)
    ax=fig.add_subplot(3,2,i+1)
    plt.hist(data=df, x=var_name, bins=bins, color=base_color)
    plt.xscale('log')
    plt.xticks([10, 20, 50, 100, 200],[10, 20, 50, 100, 200])
    ax.set_title(var_name + ' (min)')
plt.show()
```

## Delay Time by each Reason

> Of the 5 reasons, CarrierDelay, NASDelay and LateAircraftDelay are the reasons with longest delay time and WeatherDelay and SecurityDelay are the reasons that rarely happend and also have short delay time

```python

sb.barplot(data=df_reason, x='Reason', y='Time', color=base_color)
plt.ylabel('Minute')
plt.title('Delay time by reason');
```

## Delaytime by Carriers and Reasons

> For most of the carriers, CarrierDelay, NASDelay and LateAircraftDelay are the most often happened reasons. But OH is one exception when having WeatherDelay as the most often reason

```python
carriers_order=df_reason.groupby('UniqueCarrier')['ArrDelay'].mean().sort_values(ascending=False).index
reason_order=['LateAircraftDelay','NASDelay','CarrierDelay','WeatherDelay','SecurityDelay']
plt.figure(figsize = [12, 5])

sb.barplot(data = df_reason, x = 'UniqueCarrier',y='Time', hue = 'Reason',
             order = carriers_order, hue_order = reason_order)
plt.legend()
plt.xlabel('Carrier')
plt.ylabel('Minute')
plt.title('Delay time by carriers and reasons');
```

## Delay time by Reason and Month

> SecurityDelay and WeatherDelay is quite stable by time, while LateAircraftDelay and NASDelay increase in the middle of the year then fail in Sep and increase again in the end of year, CarrierDelay not follow the same pattern

```python
fig=plt.figure(figsize=[16,10])
delay_by_date=df_reason.groupby(['Month','Reason'])['Time'].mean().reset_index()
colors=sb.color_palette(n_colors=5)
color_map = dict(zip(reason_order, colors))
sb.pointplot('Month', 'Time', data=delay_by_date, hue='Reason',palette=color_map)
plt.ylabel('Minute')
plt.title('Delay time by Reason and Month')
plt.show();
```

## Delay time by Reason and Scheduled arrival time

> While LateAircraftDelay happend more in the early and late flight, NASDelay and CarrierDelay have the opposite pattern, which happen more in middle of the day.

```python
plt.figure(figsize=[16,5])

bin_center=np.arange(df_reason['CRSArrTime'].min(), df_reason['CRSArrTime'].max() +100, 100)
arrtime_binned=pd.cut(df_reason['CRSArrTime'], bin_center, include_lowest=True)
sb.set_palette('Set1')
sb.barplot( data=df_reason, x=arrtime_binned, y=df_reason['Time'],hue='Reason',
          hue_order = reason_order)

plt.xticks(np.arange(0, 24, 1),np.arange(0, 24, 1))
plt.ylabel('Minute')
plt.xlabel('Scheduled arrival time (hour)')
plt.title('Delay time by Reason and Scheduled arrival time');
```
